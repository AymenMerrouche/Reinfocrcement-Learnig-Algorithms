```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random_vs_random = np.load('tests/random_vs_random.npy')
```

```python
data = random_vs_random.cumsum(axis=0).T / np.arange(1,1001)
```

```python
sns.set()
plt.plot(np.arange(1,1001), data[0], label="Random 1")
plt.plot(np.arange(1,1001), data[1], label="Random 2")
plt.legend()
```

```python
DQN_vs_random = np.load('tests/DQN_vs_random.npy')
```

```python
data = DQN_vs_random.cumsum(axis=0).T / np.arange(1,101)
```

```python
sns.set()
plt.plot(np.arange(1,101), data[0], label ="DQN")
plt.plot(np.arange(1,101), data[1], label = "Random")
plt.legend()
```
